# Docstring

**Docstring(독스트링)**은 파이썬에서 제공하는 모듈, 함수, 클래스 또는 메소드를 정의할 때 사용되는 문자열 리터럴 값으로 된 문서화 기능이다.

```python
from docstring import MyClass, my_custom_method
```

```python
my_custom_method.__doc__
```

```python
help(my_custom_method)
```

```python
MyClass.__doc__
```

```python
help(MyClass)
```

```python
x = MyClass()

x.my_class_method.__doc__
```

```python
help(x.my_class_method)
```
